```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## Определяем данные

```python
data = np.array([-20, -60, -10, 30, 60, 70, -10, -30, -120, -100, 
                 -80, 20, 40, -60, -10, 20, 30, -80, 60, 70])
```

## Строим статистическую функцию распределения

```python
fig, ax = plt.subplots()
sns.ecdfplot(data=data, color='g')
plt.title("Статистическая функция распределения")
plt.xlabel("Углы скольжения самолета в момент сбрасывания бомбы")
plt.grid()  # Добавляем сетку
plt.xlim(-125, 80)
plt.ylim(0, 1.1)

#Строим дополнительные графики для продления основного графика
#в нуле и единице
x0 = [-130,-120]
y0 = [0, 0]
x1 = [70,90]
y1 = [1, 1]
plt.plot(x0, y0, color='g')
plt.plot(x1, y1, color='g')

plt.show()
```

## Строим гистограмму

```python
sns.histplot(data, color = 'green', bins = 8)
plt.title("Гистограмма с частотами попадания")
plt.xlabel("Углы скольжения самолета в момент сбрасывания бомбы")
plt.grid(linewidth=0.3)  # Добавляем сетку
plt.show()
```

```python
sns.histplot(data, stat="density", color = 'green', bins = 8)
plt.title("Гистограмма статистического ряда")
plt.xlabel("Углы скольжения самолета в момент сбрасывания бомбы")
plt.ylabel("Вероятность")
plt.grid(linewidth=0.3)  # Добавляем сетку
plt.show()
```
